```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

OBSERVATION-Zomato maximum records and transactions are from india then USA and then United Kingdom

```python
ratings = final_df.groupby(
    ['Aggregate rating','Rating color','Rating text']
).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

##OBSERVATION
when rating is between 4.5to 4.9 --->excellent food delivered
when the ratings are between 3.0 to 4.5---->ratings are very good
rating is between 3.5 to 3.9------->good
rating is in between 3.0 to 2.4----->average
below 2.4---->poor



```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings,hue='Rating color' ,palette=['white','red','orange','yellow','green','green']) 
```

OBSERVATIONS
not rated count is very high
maximum number of rating are between 2.5 to 3.4

```python
which country has given 0 rating
final_df[final_df['Aggregate rating'] == 0].groupby('Country').size().reset_index().rename(columns={0:'Zero Rating Count'}).sort_values('Zero Rating Count', ascending=False)   
```

Maximum number of zero rating are from Indian customers

```python
##which country has what currency
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
##which countries do have online delivery options
final_df[final_df['Has Online delivery'] == 'Yes'].Country.value_counts()
```

```python
city_counts = final_df['City'].value_counts().head(5)
```

```python
city_values = city_counts.values
city_labels = city_counts.index
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(city_values, labels=city_labels, autopct='%1.2f%%')
plt.title("Top 5 Cities with Most Restaurants")
plt.show()
```
